**1 String & List Interaction**

```python
complete_name: str ="Abdur Rehman"
letters_list = [letter for letter in complete_name if letter != " "]
print(letters_list)

modified_list = [letter.upper() if index < 3 else letter.lower() for index, letter in enumerate(letters_list)]
print(modified_list)

```
